```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# Reading data

```python
data = pd.read_excel("/kaggle/input/datasets/mahmoudalimahmoud/customer-invoice/merged_output.xlsx")
df=data.copy()
df
```

```python
df.isna().sum()
```

```python
df["Company"] = df["Company"] .fillna("unknown")
df["State"] = df["State"] .fillna("unknown")
df["PostalCode"] = df["PostalCode"] .fillna("unknown")
df["Phone"] = df["Phone"] .fillna("unknown")
df["Fax"] = df["Fax"] .fillna("unknown")
df["BillingState"] = df["BillingState"] .fillna("unknown")
df["BillingPostalCode"] = df["BillingPostalCode"] .fillna("unknown")
```

```python
df.isna().sum()
```

```python
df["year_invoice"] = df["InvoiceDate"].dt.year
df["month_invoice"] = df["InvoiceDate"].dt.month
df["day_invoice"] = df["InvoiceDate"].dt.day_name()
```

```python
df["year_invoice"].value_counts()
```

# Top 10 customers in terms of revenue?

```python
top_10_customers = df.groupby(['CustomerId','FirstName','LastName'])['Total'].sum().sort_values(ascending=False).head(10).reset_index()
top_10_customers
```

```python
fig_top_10_customers = px.bar(top_10_customers,x="CustomerId",y="Total",text_auto=True,title=("top10 customers by revenue"))
fig_top_10_customers.show()
```

# less 10 customers in terms of revenue?

```python
less_10_customers = df.groupby('CustomerId')['Total'].sum().sort_values(ascending=False).tail(10).reset_index()
less_10_customers
```

# Top 10 countries in terms of revenue?¶

```python
top_10_Country = df.groupby('BillingCountry')['Total'].sum().sort_values(ascending=False).head(10).reset_index()
top_10_Country
```

```python
fig_top_10_Country = px.bar(top_10_Country,x="BillingCountry",y="Total",title=("Top 10 countries in terms of revenue"),text_auto=True)
fig_top_10_Country.show()
```

# total revenue by year

```python
Year_revenue = df.groupby('year_invoice')['Total'].sum().reset_index()
Year_revenue
```

```python
fig_Year_revenue = px.bar(Year_revenue,x="year_invoice",y="Total",title=("total revenue by year"),text_auto=True)
fig_Year_revenue
```

# total revenue by month

```python
month_revenue = df.groupby('month_invoice')['Total'].sum().reset_index()
month_revenue
```

```python
fig_month_revenue = px.line(month_revenue,x="month_invoice",y="Total",title=("total revenue of month"),markers=True)
fig_month_revenue.show()
```

# total revenue by day_name

```python
day_revenue = df.groupby('day_invoice')['Total'].sum().reset_index()
day_revenue
```

```python
fig_day_revenue = px.line(day_revenue,x="day_invoice",y="Total",title=("total revenue of day"),markers=True)
fig_day_revenue.show()
```

# total revenue , quantity by each country

```python
country_revenue=df.groupby('BillingCountry').agg({
    'Quantity':'sum',
    'Total':'sum'
}).reset_index().sort_values(by="Total",ascending=False)
country_revenue
```

```python
fig_country_revenue = px.bar(df.groupby("BillingCountry")["Total"].sum().reset_index().sort_values(by="Total",ascending=False),x="BillingCountry",y="Total",text_auto=True,title=("total revenue by each country"))
fig_country_revenue.show()
```

```python
fig_country_quantity = px.bar(df.groupby("BillingCountry")["Quantity"].sum().reset_index().sort_values(by="Quantity",ascending=False),x="BillingCountry",y="Quantity",text_auto=True,title=("total quantity by each country "))
fig_country_quantity.show()
```
